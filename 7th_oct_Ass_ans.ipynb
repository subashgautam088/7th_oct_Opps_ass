```python
##  1


# MAking class to get the details about the bank account 

class Bank_account:
      def __init__(self,account_number,account_holder_name,balance):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

      def deposit(self,amount):  # Function to deposit money
        self.balance += amount
        print("Deposit of ",amount,"successful.")
        print("Current balance: ",self.balance)

      def withdraw(self,amount):  # Function to withdraw money
        if self.balance < amount:
          print("Insufficient balance.")
        else:
          self.balance -= amount
          print("Withdrawal of ",amount,"successful.")
          print("Current balance: ",self.balance)

      def check_balance(self):  # Function to check the balance
        print("Current balance: ",self.balance)

      def account_info(self): # Function to display the account information
        print("Account holder ",self.account_holder_name)
        print("Account number",self.account_number)
        print("Balance ",self.balance)

# Creating a new bank account
account_details = Bank_account(34567845243,"Subash Gautam",10000)

# Displaying the account information
account_details.account_info()

# Checking account detals
account_details.check_balance()

# Depositing money into the account
account_details.deposit(5000) 

# Withdrawing money from the account
account_details.withdraw(8000)
# Withdrawing money from the account
```

```python
##  2


class Employee:
  def __init__(self,employee_id,employee_name,employee_salary):
    self.employee_id = employee_id
    self.employee_name = employee_name
    self.employee_salary = employee_salary

  def display_details(self):  # Function to display the employee details
    print("Employee ID: ",self.employee_id)
    print("Employee Name: ",self.employee_name)
    print("Employee salary",self.employee_salary)

  def yearly_bonus(self,bonus_percentage): # Function to calculate the yearly bonus of employee
    bonus = (bonus_percentage / 100) * self.employee_salary
    print("The yearly bonus to the employee is ",bonus)

# Creating a new employee 
employee_bonus = Employee(1234,"Subash Gautam",10000)
employee_bonus.display_details()

# Giving 10 % bonus to the employee 
employee_bonus.yearly_bonus(10)
```

```python
##  3


class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}

    def add_vehicle(self, vehicle_id, vehicle_type):
        self.available_vehicles[vehicle_id] = vehicle_type
        print(f"Vehicle {vehicle_id} ({vehicle_type}) added to the rental system.")

    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            vehicle_type = self.available_vehicles.pop(vehicle_id)
            print(f"Vehicle {vehicle_id} ({vehicle_type}) has been rented.")
        else:
            print(f"Vehicle {vehicle_id} is not available for rent.")

    def return_vehicle(self, vehicle_id, vehicle_type):
        self.available_vehicles[vehicle_id] = vehicle_type
        print(f"Vehicle {vehicle_id} ({vehicle_type}) has been returned.")

    def display_available_vehicles(self):
        if self.available_vehicles:
            print("Available Vehicles:")
            for vehicle_id, vehicle_type in self.available_vehicles.items():
                print(f"Vehicle ID: {vehicle_id}, Type: {vehicle_type}")
        else:
            print("No vehicles available for rent.")


# Example Usage
if __name__ == "__main__":
    # Creating a vehicle rental system
    rental_system = VehicleRentalSystem()

    # Adding vehicles to the system
    rental_system.add_vehicle("V001", "Sedan")
    rental_system.add_vehicle("V002", "SUV")
    rental_system.add_vehicle("V003", "Truck")

    # Displaying available vehicles
    rental_system.display_available_vehicles()

    # Renting a vehicle
    rental_system.rent_vehicle("V001")

    # Displaying available vehicles after renting
    rental_system.display_available_vehicles()

    # Returning a vehicle
    rental_system.return_vehicle("V001", "Sedan")

    # Displaying available vehicles after returning
    rental_system.display_available_vehicles()
```

```python
##  4
class Library:
  def __init__(self):
    self.books = {}

  def add_book(self, book_id, book_name, author_name):
    self.books[book_id] = {
      "book_id": book_id,
      "book_name": book_name,
      "author_name": author_name,
      "available": True
    }
    print(f"Book {book_name} by {author_name} has been added to the library.")

  def borrow_book(self, book_id):
    if book_id in self.books:
      book = self.books[book_id]
      if book["available"]:
        book["available"] = False
        print(f"Book {book['book_name']} by {book['author_name']} has been borrowed.")

  def return_book(self, book_id):
    if book_id in self.books:
      book = self.books[book_id]
      if not book["available"]:
        book["available"] = True
        print(f"Book {book['book_name']} by {book['author_name']} has been returned.")

  def display_books(self):
    if self.books:
      print("Available Books:")
      for book_id, book in self.books.items():
        if book["available"]:
          print(f"Book ID: {book_id}, Book Name: {book['book_name']}, Author Name: {book['author_name']}")
    else:
      print("No books available in the library.")

# Example Usage
if __name__ == "__main__":
  # Creating a library
  library = Library()

  # Adding books to the library
  library.add_book("B001", "The Great Gatsby", "F. Scott Fitzgerald")
  library.add_book("B002", "To Kill a Mockingbird", "Harper Lee")
  library.add_book("B003", "1984", "George Orwell")

  # Displaying available books
  library.display_books()

  # Borrowing a book
  library.borrow_book("B001")

  # Displaying available books after borrowing
  library.display_books()
```

```python
##  5


class Product:
  def __init__(self, product_id, name, price, quantity):
      self.product_id = product_id
      self.name = name
      self.price = price
      self.quantity = quantity

  def update_quantity(self, new_quantity):
      self.quantity = new_quantity


class Inventory:
  def __init__(self):
      self.products = {}

  def add_product(self, product):
      if product.product_id not in self.products:
          self.products[product.product_id] = product
      else:
          existing_product = self.products[product.product_id]
          existing_product.update_quantity(existing_product.quantity + product.quantity)

  def update_product_quantity(self, product_id, new_quantity):
      if product_id in self.products:
          self.products[product_id].update_quantity(new_quantity)
      else:
          print(f"Product with ID {product_id} not found in the inventory.")

  def display_available_products(self):
      print("Available Products:")
      for product_id, product in self.products.items():
          print(f"ID: {product_id}, Name: {product.name}, Price: {product.price}, Quantity: {product.quantity}")


# Example usage
if __name__ == "__main__":
  # Creating products
  product1 = Product(1, "Laptop", 800, 10)
  product2 = Product(2, "Mouse", 20, 50)
  product3 = Product(3, "Keyboard", 50, 30)

  # Creating an inventory system
  inventory = Inventory()

  # Adding products to the inventory
  inventory.add_product(product1)
  inventory.add_product(product2)
  inventory.add_product(product3)

  # Displaying available products
  inventory.display_available_products()

  # Updating product quantity
  inventory.update_product_quantity(1, 5)  # Updating Laptop quantity to 5

  # Displaying available products after updating quantity
  inventory.display_available_products()
```

```python
##   6

class Shape:
  def __init__(self,length,width,height):
    self.l = length
    self.w = width
    self.h = height
    
  def calculate_area(self):
      print("Choose shape from this Rectangle,Traingle,Square")
      shapes_list = input("Enter shape to calculate area :")
      if shapes_list == "Rectangle":
          print ("Area of rectangle is ",self.l * self.w)
      elif shapes_list =="Traingle":
         print(f"Area of traingle is {(self.l * self.w) / 2}")
      elif shapes_list =="Square" :
          print("Area of square is :", (self.l ** 2))
      else:
          print("Invalid shape")
        
  def calculate_perimeter(self):
      shapes_list = input("Enter shape to calculate perimeter :")
      if shapes_list == "Rectangle":
          print ("Perimeter of rectangle is ",2 * (self.l + self.w))
      elif shapes_list =="Traingle":
         print(f"Perimeter of traingle is {self.l + self.w + self.h}")
      elif shapes_list =="Square":
          print("Perimeter of square is :", 4 * self.l)
      else:
          print("Invalid shape")



obj2 = Shape(10,20,30)
obj2.calculate_area()
obj2.calculate_perimeter()
```

```python
#  7


class Student:
  def __init__(self,student_Id,student_name):
    self.Id = student_Id
    self.name = student_name
    self.grade = []

  def add_grade(self,marks):  # Add marks 
    self.grade.append(marks)

  def calculate_average_grade(self): #Calclate average grade of student
    avg_grade = sum(self.grade)/len(self.grade)
    if 90<avg_grade<=100:
      return f"A ,{avg_grade}Percentage"
    if 80<avg_grade<=90:
      return f"B ,{avg_grade}Percentage"
    if 70<avg_grade<=80:
      return f"C ,{avg_grade}Percentage"
    if 60<avg_grade<=70:
      return f"D ,{avg_grade}Percentage"
    if 50<avg_grade<=60:
      return f"E ,{avg_grade}Percentage"

  def dispaly_details(self):    #Display student details.
    print(f"Student Id ={self.Id}")
    print(f"Name ={self.name}")
    print(f"Grades ={self.grade}")
    print(f"AVerage Grade ic {self.calculate_average_grade()}")


student1 = Student(4,"Alice")
student1.add_grade(76)
student1.add_grade(84)
student1.add_grade(59)
student1.add_grade(99)
student1.add_grade(59)
student1.calculate_average_grade()
student1.dispaly_details()
```

```python
##  8


class Email:
  def __init__(self, sender, recipient, subject, message):
      self.sender = sender
      self.recipient = recipient
      self.subject = subject
      self.message = message
      self.a="Myself"

  def send_email(self):
      # We will just print the details.
      print(f"Email sent from: {self.sender}")
      print(f"Email sent to: {self.recipient}")
      print(f"Subject: {self.subject}")
      print("Message:")
      print(self.message)
      print("Email sent successfully.")

  def display_email_details(self):
      print(f"Sender: {self.sender}")
      print(f"Recipient: {self.recipient}")
      print(f"Subject: {self.subject}")
      print("Message:")
      print(self.message)


# Example usage
if __name__ == "__main__":
  # Creating an email
  sender = "sender@example.com"
  recipient = "recipient@example.com"
  subject = "Sample Email"
  message = "Hello, this is a sample of pwskills email message."

  email = Email(sender, recipient, subject, message)

  # Displaying email details
  print("Email Details:")
  email.display_email_details()

  # Sending the email
  print("\nSending Email:")
  email.send_email()
```

```python
##  9


class SocialMediaProfile:
  def __init__(self,username):
    self.username = username
    self.post = []

  def add_post(self,posts):
    self.post.append(posts)

  def display_posts(self):
    print(f"Post by {self.username}:")
    for idx, post in enumerate(self.post, start=1):
        print(f"{idx}. {post}")

  def search_post_by_keyword(self,keyword):
    matching_posts = [post for post in self.post if keyword in post]
    print(f"Posts containing '{keyword}' are as follows:")
    if matching_posts:
        for idx, post in enumerate(matching_posts, start=1):
            print(f"{idx}. {post}")
    else:
        print("No matching posts found.")


if __name__=="__main__":
  user ="123user"
  # Making an object of SocialMediaProfile Class.
  profile = SocialMediaProfile(user)

  # Adding post one by one.
  profile.add_post("It is my 1st post.")
  profile.add_post("It's a second post from my side.")
  profile.add_post("Hey yarro its a 3rd post.")
  profile.add_post("Its a pwskills online platform to study.")

  profile.display_posts()  #Display a details about post.

  # Searching for posts containing a specific keyword
  profile.search_post_by_keyword("study")
```

```python
##  10


class ToDoList:
  def __init__(self):
      self.tasks = []

  def add_task(self, task, due_date):
      self.tasks.append({"task": task, "due_date": due_date, "completed": False})

  def mark_task_as_completed(self, task_index):
      if 0 <= task_index < len(self.tasks):
          self.tasks[task_index]["completed"] = True
      else:
          print("Invalid task index.")

  def display_pending_tasks(self):
      print("Pending Tasks:")
      for idx, task in enumerate(self.tasks):
          if not task["completed"]:
              print(f"{idx}. Task: {task['task']}, Due Date: {task['due_date']}")

# Example usage
if __name__ == "__main__":
  # Creating a ToDo list
  todo_list = ToDoList()

  # Adding tasks to the ToDo list
  todo_list.add_task("Eating breakfast", "2023-11-15")
  todo_list.add_task("Exercise", "2023-11-10")
  todo_list.add_task("Doing Homework", "2023-11-12")

  # Marking a task as completed
  todo_list.mark_task_as_completed(0)  # Marking "Eating breakfast" task as completed

  # Displaying pending tasks
  todo_list.display_pending_tasks()
```
